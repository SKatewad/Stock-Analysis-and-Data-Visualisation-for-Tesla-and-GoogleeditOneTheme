```python
import pandas as pd
```

```python
Google =pd.read_excel("GOOG_data.xlsx")
Tesla=pd.read_excel("Tesla_data.xlsx")
```

```python
Google.head()
```

```python
Tesla.head()
```

```python
dfs = [Google,Tesla]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
Tesla.head()
```

```python
for df in dfs:
    df['Previous day close price']= df.Close.shift(1)
```

```python
Google.head()
```

```python
for df in dfs:
    df['Change in price'] = df['Close'] - df['Previous day close price']
```

```python
Google.head()
```

```python
for df in dfs:
 df['Percent change in price'] = df.Close.pct_change()
```

```python
Google.head()
```

```python
for df in dfs: 
 df['Previous day volume']= df.Volume.shift(1)
```

```python
Google.head()
```

```python
for df in dfs:
 df['Change in volume'] = df['Volume'] - df['Previous day volume']
```

```python
Google.head()
```

```python
for df in dfs:
    df['Percent change in volume']= df.Volume.pct_change()
```
